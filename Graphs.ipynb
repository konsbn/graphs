##Cardioid
This is one of the first curves I discovered when I was reading a book on differential calculus. It was an old book and was full of curve tracing. 
Cardioid as you might have guessed has it's roots in the greek word for heart.
The equation is 
>## $$r = a + b*cos(\theta)$$ 

where

> ## $$\theta \in [\ 0, 2 \pi]\ $$
   

Of course it is plotted as a Polar Graph. If you wanna read up more on Cardioid and I suggest you should head over to the [wikipedia page]
[wikipedia page]:http://en.wikipedia.org/wiki/Cardioid



##Getting Started you'll need to start the Ipython Shell and import these


```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import *
```

Now that you have the basic setup let's start with plotting the cardioid.
starting with setting up values of 
## $$\theta$$

```python
theta = arange(0.0, 2*pi , 0.1)
```

now for the 
>## $$a \& b$$

```python
a = 1.0 
b = 1.0
```

## $$r = $$

```python
r = map(lambda x:a + b*cos(x),theta)
```

>Now we plot it as a polar plot

```python
plt.polar(theta, r)
```
